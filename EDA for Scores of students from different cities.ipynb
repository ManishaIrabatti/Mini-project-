```python
names=list(data['Name'])
gender=list(data["Gender"])
city=list(data['CityTown'])
maths=list(data['Mathematics'])
physics=list(data['Physics'])
chem=list(data['Chemistry'])
total=list(data['Total'])
```

```python
#Precentage of students by cities
import matplotlib.pyplot as plt
cityt=list(dict.fromkeys(city))  #keys nd
stud_pcity=[]
for i in cityt:
    cn=city.count(i)
    stud_pcity.append(cn)
stud_pcity                     #no of students from each city
prcnt=[]
for i in stud_pcity:
    prc=round((i/(sum(stud_pcity))*100),2)
    prcnt.append(prc) 
prcnt                             # percentage of students from diff cities
fig=plt.figure(figsize=(10,5))  
plt.pie(stud_pcity,labels=prcnt)

plt.title('Percentage of students from each City')   
plt.legend(cityt)

plt.show  
```

```python
#genderwise percentage
cntm=0
cntf=0
for i in gender:
    if i=='M':
        cntm+=1
    elif i=='F':
        cntf+=1
        
gndr_count={}
gndr_count['Male']=cntm
gndr_count['Female']=cntf
gndr_count                           #male17 and female13 count
gndr_percnt=[]
for i in gndr_count.values():
    per=round((i/sum(gndr_count.values())*100),2)
    gndr_percnt.append(per)
gndr_percnt                           #male 56.67 female 43.32

fig=plt.figure(figsize=(10,5))   
plt.pie(gndr_count.values(),labels=gndr_percnt)

plt.title('Percentage male and female students')   
plt.legend(gndr_count.keys())

plt.show 
```

```python
#common functions
def median(Data):
    Data.sort()
    N=int(len(Data))
    if N%2==0:
        median=(Data[int(N/2)-1]+Data[(int(N/2)+1)-1])/2
    else:
        median=Data[int((N+1)/2)-1]
    return median

def mean(Data):
    N=int(len(Data))
    summ=0
    for i in Data:
        summ=i+summ
    mean=summ/N
    return mean

def mode(Data):
    count_dic={}
    for i in Data:
        c=Data.count(i)
        count_dic[i]=c
    mode=max(count_dic,key=count_dic.get)
    return mode
def var(Data):
    N=int(len(Data))
    diff=[]
    for i in Data:
        diff_1=(i-mean(Data))**2
        diff.append(diff_1)
    summ_sq_diff=0
    for i in range(N):
        summ_sq_diff+=diff[i]
    var=summ_sq_diff/N
    return var

def std_dev(Data):
    std_dev=(var(Data))**0.5
    return std_dev
def rng(Data):
    R=max(Data)-min(Data)
    return R
```
